```python
# baca data reku_selected3.csv

import pandas as pd

df = pd.read_csv('jobstreet_3000_selected3.csv')
print(df.head())
```

```python
corpus = df['content']
print(corpus.head())
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from sklearn.feature_extraction.text import TfidfVectorizer

# Membuat daftar stopwords dari Sastrawi
factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()  # Mengambil daftar stopwords sebagai list


vectorizer = TfidfVectorizer(stop_words=stopwords)
response = vectorizer.fit_transform(corpus)
print(response)
```

```python
vectorizer.get_feature_names_out()
```

```python
response.todense()
```

```python
import pandas as pd

df = pd.DataFrame(response.todense().T,
                  index=vectorizer.get_feature_names_out(),
                  columns=[f'D{i+1}' for i in range(len(corpus))])

df
```
