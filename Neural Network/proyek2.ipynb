# Prediksi Diabetes dengan FFNN

```python
import pandas as pd

# Load the CSV file into a DataFrame
data = pd.read_csv('sample_data/diabetes.csv')

# Convert the DataFrame to a 2-dimensional array
data_array = data.values

# Display the array
print(data_array)
data.head()
```

```python
# Pisahkan kolom Outcome menjadi array dua dimensi
outcome_array = data[['Outcome']].values
data_array = data.drop('Outcome', axis=1).values
# Tampilkan array Outcome
print(outcome_array)
print(data_array)
```

```python
%pip install tensorflow

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np

model = Sequential([
    Dense(4, input_dim=data_array.shape[1], activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(data_array, outcome_array, epochs=2000, verbose=1)

# predictions = model.predict()
# print("Prediksi:")
# print(predictions)
```

## Predict

```python
a = np.array([[  6,    148,     72,    35, 0, 33.6,  0.627,  50],
	[5, 121, 72, 23, 0, 26.2, 0.245, 30],
	[1, 126, 60, 23, 0, 30.1, 0.349, 47],
	[1, 93, 70, 23, 0, 30.4, 0.315, 23]
])

predictions = model.predict(a)
print("Prediksi:")
print(predictions)
```

## Labeling

```python
import numpy as np
pred_labels = (predictions > 0.5).astype(int)
print(pred_labels)
```
