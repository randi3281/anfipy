# Menjadikan kata menjadi token atau menghitung jumlah kata yang ada di dalam data berdasarakan vectorized

```python
from sklearn.feature_extraction.text import CountVectorizer

X = data['content']
y = data['score']

vectorized = CountVectorizer()
vectorized_X = vectorized.fit_transform(X).todense()
vectorized_X
```

# Memunculkan Dictionary Kata atau Token

```python
vectorized.get_feature_names_out()
```

# Euclidiean Distance (Mengukur Jarak Dokumen 1 dengan lainnya)

```python
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

# Loop untuk menghitung jarak Euclidean antar dokumen
for i in range(len(vectorized_X)):
    for j in range(i + 1, len(vectorized_X)):
        # Pastikan tipe data adalah np.array
        doc_i = np.asarray(vectorized_X[i]).reshape(1, -1)
        doc_j = np.asarray(vectorized_X[j]).reshape(1, -1)

        jarak = euclidean_distances(doc_i, doc_j)
        print(f'Jarak Dokumen {i+1} dan Dokumen {j+1} adalah {jarak[0][0]}')
```

# Stop Word Processing

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from sklearn.feature_extraction.text import CountVectorizer

# Membuat daftar stopwords dari Sastrawi
factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()  # Mengambil daftar stopwords sebagai list

# Menggunakan CountVectorizer dengan stopwords dari Sastrawi
vectorized2 = CountVectorizer(stop_words=stopwords)

# Melakukan fit-transform pada data X
vectorized_X2 = vectorized2.fit_transform(X).todense()

# Menampilkan hasil
vectorized_X2
```

# Euclidiean Distance (Mengukur Jarak Dokumen 1 dengan lainnya)

```python
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

# Loop untuk menghitung jarak Euclidean antar dokumen
for i in range(len(vectorized_X2)):
    for j in range(i + 1, len(vectorized_X2)):
        # Pastikan tipe data adalah np.array
        doc_i = np.asarray(vectorized_X2[i]).reshape(1, -1)
        doc_j = np.asarray(vectorized_X2[j]).reshape(1, -1)

        jarak = euclidean_distances(doc_i, doc_j)
        print(f'Jarak Dokumen {i+1} dan Dokumen {j+1} adalah {jarak[0][0]}')
```

# Menampilkan Jarak Antar Dokumen dari yang Terdekat

```python
# urutkan jarak yang terdekat
jarak = []
for i in range(len(vectorized_X2)):
    for j in range(i + 1, len(vectorized_X2)):
        doc_i = np.asarray(vectorized_X2[i]).reshape(1, -1)
        doc_j = np.asarray(vectorized_X2[j]).reshape(1, -1)

        jarak.append((i, j, euclidean_distances(doc_i, doc_j)[0][0]))

# Urutkan jarak dari yang terkecil dan tampilkan jaraknya dalam np.float64
jarak = sorted(jarak, key=lambda x: x[2])
jarak[:10]
```

# Menampilkan 10 Jarak Antar Dokumen dari yang Terdekat

```python
# Urutkan jarak dari yang terbesar dan tampilkan jaraknya dalam np.float64
jarak = sorted(jarak, key=lambda x: x[2], reverse=True)
jarak[:10]
```

# Menampilkan Jarak Antar Dokumen dari yang Terjauh

```python
# Tampilkan semua dokumen yang memiliki jarak terdekat
for i, j, jarak in jarak[:10]:
    print(f'Dokumen {i+1} dan Dokumen {j+1} memiliki jarak {jarak}')
    print(f'Dokumen {i+1}: {X[i]}')
```
